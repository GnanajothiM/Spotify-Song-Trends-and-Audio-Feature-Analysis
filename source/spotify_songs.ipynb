Which music genre has the highest average popularity?

```python
spotify.groupby("playlist_genre")["track_popularity"].mean().sort_values(ascending=False)
```

Which song has the highest popularity score?

```python
spotify.sort_values("track_popularity", ascending=False )[["track_name","track_popularity"]]
```

Which artist has released the most songs?

```python
most_songs=spotify.groupby("track_artist")["track_name"].count().sort_values(ascending=False).head(10)
```

```python
most_songs.plot(kind="bar")
plt.ylabel("Number of Songs")
plt.title("Top 10 Artists with Most Songs")
plt.show()
```

Which genre contains the highest number of songs?

```python
high=spotify.groupby("playlist_genre")["track_name"].count()
```

```python
high.plot(kind="bar")
plt.ylabel("count of songs")
plt.title("Number of Songs by Genre")
plt.show()
```

Which album has the highest total popularity?

```python
total=spotify.groupby("track_album_name")["track_popularity"].sum().sort_values(ascending=False).head(10)   
```

```python
total.plot(kind="bar")
```

Which language has the highest number of songs?

```python
ls.plot(kind="pie")
```

Which artist has the highest total popularity?

Songs Released Per Year

```python
spotify["Year"] = spotify["track_album_release_date"].dt.year
```

```python
yearly=spotify.groupby("Year")["track_name"].count()
```

```python
yearly.plot(kind="line")
plt.ylabel("count of songs")
plt.title("Number of Songs Released by Year")
plt.show()
```
